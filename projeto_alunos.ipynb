```python
nomes = df_alunos.query('Nome == "Alice" | Nome == "Carlos"').index
nomes
```

```python
aprovado = df_alunos['Aprovado'] == True
df_alunos[aprovado]
```

```python
alunos_aprovados = df_alunos[aprovado]
```

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', sep = ';', index=False)
```

```python
pd.read_csv('alunos_aprovados.csv', sep=';')
```

```python
alunos_aprovados = alunos_aprovados.replace(7.0, 8.0)
```

```python
alunos_aprovados
```

```python
df_alunos['Pontos_extras'] = df_alunos['Notas'].apply(lambda x: x * 0.4)
df_alunos
```

```python
df_alunos['Notas_finais'] = df_alunos['Notas'] + df_alunos['Pontos_extras']
df_alunos
```

```python
df_alunos['Aprovado_final'] = df_alunos['Notas_finais'].apply(lambda x: True if x >= 6 else False)
df_alunos
```

```python
df_alunos.query('Aprovado == False & Aprovado_final == True')
```
